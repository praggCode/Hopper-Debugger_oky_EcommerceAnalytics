# Notebook 01 — Data Extraction & Initial Inspection
**Project:** Olist E-Commerce Performance Analysis  
**Purpose:** Load all raw CSV files, inspect structure, document shape, dtypes, and null counts  
**Output:** Summary inspection saved to docs — no data is modified here


```python
import pandas as pd
import numpy as np

print("Libraries loaded successfully")
```

```python
orders       = pd.read_csv("../data/raw/olist_orders_dataset.csv")
order_items  = pd.read_csv("../data/raw/olist_order_items_dataset.csv")
customers    = pd.read_csv("../data/raw/olist_customers_dataset.csv")
payments     = pd.read_csv("../data/raw/olist_order_payments_dataset.csv")
reviews      = pd.read_csv("../data/raw/olist_order_reviews_dataset.csv")
products     = pd.read_csv("../data/raw/olist_products_dataset.csv")
sellers      = pd.read_csv("../data/raw/olist_sellers_dataset.csv")
geolocation  = pd.read_csv("../data/raw/olist_geolocation_dataset.csv")
category_translation = pd.read_csv("../data/raw/product_category_name_translation.csv")

print("All 9 tables loaded successfully")
```

```python
tables = {
    "orders": orders,
    "order_items": order_items,
    "customers": customers,
    "payments": payments,
    "reviews": reviews,
    "products": products,
    "sellers": sellers,
    "geolocation": geolocation,
    "category_translation": category_translation
}

print(f"{'Table':<25} {'Rows':>10} {'Columns':>10}")
print("-" * 47)
for name, df in tables.items():
    print(f"{name:<25} {df.shape[0]:>10,} {df.shape[1]:>10}")
```

```python
print("=== SAMPLE DATA PREVIEW ===\n")
for name, df in tables.items():
    print(f"--- {name} ---")
    display(df.head(3))
    print()
```

```python
print("=== DATAFRAME INFO ===\n")
for name, df in tables.items():
    print(f"--- {name} ---")
    df.info()
    print()
```

```python
print("=== STATISTICAL SUMMARY (Numerical Columns) ===\n")
for name, df in tables.items():
    numeric = df.select_dtypes(include='number')
    if not numeric.empty:
        print(f"--- {name} ---")
        display(numeric.describe().round(2))
        print()
```

```python
print("=== NULL VALUE REPORT ===\n")
for name, df in tables.items():
    null_counts = df.isnull().sum()
    nulls = null_counts[null_counts > 0]
    print(f"--- {name} ---")
    if nulls.empty:
        print("  No null values found")
    else:
        for col, count in nulls.items():
            pct = (count / len(df)) * 100
            print(f"  {col}: {count:,} nulls ({pct:.1f}%)")
    print()
```

```python
print("=== DATA TYPES REPORT ===\n")
for name, df in tables.items():
    print(f"--- {name} ---")
    print(df.dtypes.to_string())
    print()
```

```python
print("=== DUPLICATE ROWS REPORT ===\n")
for name, df in tables.items():
    dupes = df.duplicated().sum()
    print(f"  {name}: {dupes:,} duplicate rows")
```

```python
print("=== UNIQUE VALUE CHECK ===\n")
print("orders.order_id unique:", orders['order_id'].nunique())
print("customers.customer_unique_id unique:", customers['customer_unique_id'].nunique())
```

```python
print("=== ORDER STATUS DISTRIBUTION ===\n")
print(orders['order_status'].value_counts())
print(f"\nTotal orders: {len(orders):,}")
```

```python
print("=== REVIEW SCORE DISTRIBUTION ===\n")
print(reviews['review_score'].value_counts().sort_index())
```

```python
print("=== PAYMENT TYPE DISTRIBUTION ===\n")
print(payments['payment_type'].value_counts())

print("\n=== INVALID INSTALLMENTS (value = 0) ===")
print(payments[payments['payment_installments'] == 0].shape[0], "rows with 0 installments")
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

print("=== ORDER DATE RANGE ===")
print(f"  Earliest order: {orders['order_purchase_timestamp'].min()}")
print(f"  Latest order:   {orders['order_purchase_timestamp'].max()}")
```

## Extraction Summary

All 9 raw tables loaded and inspected successfully.
No data was modified in this notebook.

Key issues identified for cleaning:
- Geolocation: 261,831 duplicate rows — deduplicate by zip code
- Orders: nulls in delivery timestamps — filter by order_status
- Payments: 2 invalid installment values + 3 not_defined payment types
- Products: 610 rows with missing category and dimensions
- Reviews: 88% null comment fields — drop comment columns
- Customers: 96,096 unique customers across 99,441 orders — 3,345 repeat buyers identified
